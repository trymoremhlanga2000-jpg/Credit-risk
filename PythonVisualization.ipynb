```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
import plotly.express as px
```

```python
df = pd.read_csv("C:/Users/TRYMORE/Desktop/DataSets/creditdata.csv")
```

```python
df.columns
```

```python
#Histogram

plt.hist(df['Age'], bins = 30, color = 'green', alpha = 0.8)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Histogram')
plt.show
```

```python
#Bar Plot
plt.bar(df['Gender'], df['Age'], color = 'green', alpha = 0.7)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Age vs Gender')
plt.show
```

```python

#Seaborn Histogram

sbn.histplot(data = df['Age'], kde = True, bins = 10, color = 'green', alpha = 0.8)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Histogram')
plt.show
```

```python
#Seaborn Barplot
sbn.barplot(x = "Gender", y = "Age", data = df, color = 'orange')
plt.show
```

```python
sbn.boxplot(x = 'Marital_Status', y = 'Age', data = df, color = 'green')
plt.show
```

```python
#Perform some joint plots

sbn.jointplot(x = 'Age', y ='Number_of_Dependants', data = df, kind = 'kde')
```

```python

sbn.jointplot(x = 'Age', y ='Number_of_Dependants', data = df, kind = 'reg')
```

```python
sbn.jointplot(x = 'Age', y ='Number_of_Dependants', data = df, kind = 'kde', shade = True)
```

```python


sbn.jointplot(x = 'Age', y ='Number_of_Dependants', data = df, kind = 'hex')
```

```python
sbn.heatmap(df.corr(), annot = True, cmap = 'Y1GnBu')   #This only works when the dataset has all continuous variables
```

```python
#plotly for interactive plots
import plotly.io as pio
pio.renderers.default = 'notebook_connected'

import plotly.express as px
fig = px.bar(df, x = "Gender", y = 'Age', title = 'Interactive Chart')
fig.show()
```

```python
fig = px.scatter_3d(df, x = 'Gender', y = 'Age', z = 'Number_of_Dependants', title = '3D Scatter')
fig.show
```

```python
#Creating Heatmaps
#Matplotlib

import numpy as np
data = np.random.rand(5,5)
plt.imshow(data, cmap = 'viridis')
plt.colorbar()
plt.title('Heatmap')
plt.show
```

```python
#Heatmap with seaborn
sbn.heatmap(data, cmap = 'viridis')

plt.title('Heatmap')
plt.show
```

```python
#Heatmap using plotly
fig = px.imshow(data, color_continuous_scale = 'viridis')
fig.show
```
