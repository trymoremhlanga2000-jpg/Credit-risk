```python
#All the clients not with employments status as...

df[~df['Employment_Status'].isin(['Employed', 'Salaried Employee', 'Unemployed'])]
```

```python
#This deletes row 38 from the dataframe
df = df.drop(38)
```

```python
#Drop/deletes anything that contains any null values or na values
df.dropna()
```

```python
#Fill all null or na values in ID with zero
df.fillna({'ID': 0}, inplace = True)
```
